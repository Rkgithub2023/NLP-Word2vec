**Word embedding technique**

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
sent=['the glass of milk',
      'the glass of jiuce',
      'the cup of tea',
      'im a very good boy',
      'im a good developer',
      'your voice is good']
```

```python
sent
```

```python
voc_size=12000
```

one hot representation

```python
onehot_rep=[one_hot(words,voc_size) for words in sent]
print(onehot_rep)
```

Word embedding representation

```python
from tensorflow.keras.layers import Embedding,InputLayer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8

embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

Layers

```python
model=Sequential()
#model.add(InputLayer(input_shape=(8,)))
model.add(Embedding(voc_size,100)) #here it will create voc_size (10000 words) into 10 dimension feature ,10 is feature dimentsion
model.compile('adam','mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])
#print(model.predict(embedded_docs[0]))
```

```python
sent=['The world is a better place',
      'Marvel series is my favourite movie',
      'I like DC movies',
      'the cat is eating the food',
      'Tom and Jerry is my favourite movie']
```

```python
vocabulary_=5000
```

```python
#converting words to onehot encoding
onehot_newreq=[one_hot(words,vocabulary_) for words in sent]
print(onehot_newreq)
```

```python
#need to do padding to bring in same size
step_size=10
paddingtip=pad_sequences(onehot_newreq,maxlen=step_size,padding='pre')
paddingtip
```
